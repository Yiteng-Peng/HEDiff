```python
import pickle
import matplotlib.pyplot as plt
```

```python
pkl_filename = f"./corpus/ts_mnist.pkl"
# pkl_filename = f"./corpus/helayer_digits.pkl"
# pkl_filename = f"./corpus/zama_mnist.pkl"

with open(pkl_filename, 'rb') as fp:
    result_tuple = pickle.load(fp)

sameList, muSameList, oriTrueDiffList, muTrueDiffList, patTrueDiffList, patternDict, seed_pattern_idx_list, mutationList = result_tuple
```

```python
start_idx = 0
row_num = 10

fig, axes = plt.subplots(3, row_num, figsize=(row_num, 3))
plt.subplots_adjust(wspace=0.05, hspace=0.25)

for i in range(row_num):
    data, _, label_p, _ = patternDict[start_idx + i]
    axes[0, i].imshow(data.squeeze().numpy(), cmap='gray')
    axes[0, i].axis('off')

    axes[0, i].text(0.5, -0.2, int(label_p), ha='center', transform=axes[0, i].transAxes)

for i in range(row_num):
    data, noise, _, label_n = patternDict[start_idx + i]
    axes[1, i].imshow((data+noise).squeeze().numpy(), cmap='gray')
    axes[1, i].axis('off')

    axes[1, i].text(0.5, -0.2, int(label_n), ha='center', transform=axes[1, i].transAxes)

for i in range(row_num):
    _, noise, _, _ = patternDict[start_idx + i]
    axes[2, i].imshow(noise.squeeze().numpy(), cmap='gray')
    axes[2, i].axis('off')

plt.show()
```
